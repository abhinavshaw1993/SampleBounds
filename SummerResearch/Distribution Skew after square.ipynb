```python
import numpy as np
import seaborn as sns
from scipy.stats import skew
from scipy.stats.distributions import truncexpon
from scipy.stats.distributions import uniform
%matplotlib inline

N = 100
samples = np.random.normal(loc=0.5, scale=0.1, size=N )
sns.distplot(samples, bins = 10)
sns.distplot(np.square(samples), bins = 10)
```

```python
# truncexpon
expon_samples = truncexpon.rvs(b=0.1, loc=0.5, scale=0.3, size=N)

sns.distplot(expon_samples, bins = 10)
sns.distplot(np.square(expon_samples), bins = 10)
```

```python
# Uniform
expon_samples = uniform.rvs(0, 1, size=N)

sns.distplot(expon_samples, bins = 10)
sns.distplot(np.square(expon_samples), bins = 10)


```

```python
# Emperical CDF

from statsmodels.distributions.empirical_distribution import ECDF

ecdf = ECDF(expon_samples)

print(expon_samples)
print (ecdf(np.linspace(0, 1, N)) )





```
