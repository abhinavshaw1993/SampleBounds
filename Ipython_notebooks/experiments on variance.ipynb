```python
from scipy.stats import norm
import seaborn as sns

%matplotlib inline

r = norm.rvs(size=1000)
sns.distplot(r, bins=20)


```

```python
import matplotlib.pyplot as plt
import numpy as np

stdev = .1
F_x = norm.cdf(np.linspace(-10,10, 100), loc=0, scale=stdev)
plt.plot(F_x, label = str(stdev))


stdev = 4
F_x = norm.cdf(np.linspace(-10,10, 100), loc=0, scale=stdev)
plt.plot(F_x, label= str(stdev))
plt.legend()


```
