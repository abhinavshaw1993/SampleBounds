```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import truncnorm
%matplotlib inline

left_support, right_support = 15.0, 20.0
mean = 17.5
std = 0.5
N = 1000

a, b = (left_support - mean) / std, (right_support - mean) / std

samples = truncnorm.rvs(a=a, b=b, loc=mean, scale=std, size=N)
ord_stats = np.sort(samples)

sns.distplot(samples, bins=10)
```

```python
print "mean of samples: ", np.mean(samples)
```

```python
samples_cdf = truncnorm.cdf(samples, a=a, b=b, loc=mean, scale=std)
plt.plot(range(N), np.sort(samples_cdf))
```

```python
dx = (right_support - left_support) / N

cdf_integral = 0.0

for i in range(1, N+1):
    x0 = left_support + (i-1)*dx
    x1 = left_support + i*dx
    cdf_x0 = truncnorm.cdf(x0, a, b, loc=mean, scale=std)
    cdf_x1 = truncnorm.cdf(x1, a, b, loc=mean, scale=std)
    cdf_integral += ((cdf_x0 + cdf_x1)/2.0)*dx
    
computed_mean = right_support - cdf_integral
print "cdf integral: ", cdf_integral
print "computed mean: ", computed_mean
```
