```python
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")

# Load the long-form example gammas dataset
gammas = sns.load_dataset("gammas")

# Plot the response with standard error
sns.tsplot(data=gammas, time="timepoint", unit="subject",
           condition="ROI", value="BOLD signal")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")

values = np.array(
    [
        [1, 0, 1.0, 1],
        [1, 1, 1.5, 1],
        [1, 2, 2.0, 1],
        [1, 3, 1.75, 1],
        [1, 0, -1.0, 2],
        [1, 1, -1.5, 2],
        [1, 2, -2.0, 2],
        [2, 0, 1.0, 1],
        [2, 1, 1.5, 1],
        [2, 2, 2.0, 1],
        [2, 3, 1.75, 1],
        [2, 0, -1.0, 2],
        [2, 1, -1.5, 2],
        [2, 2, -2.0, 2],
        [3, 0, 1.0, 1],
        [3, 1, 1.5, 1],
        [3, 2, 2.0, 1],
        [3, 3, 1.75, 1],
        [3, 0, -1.0, 2],
        [3, 1, -1.5, 2],
        [3, 2, -2.0, 2]
    ]
)

df = pd.DataFrame(data=values, columns=['N', 'OBS', 'Bound', 'Type'])
# print df

sns.tsplot(data=df, time="N", unit="OBS",
           condition="Type", value="Bound", ci=95)
```
